```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/spotify_music.csv")
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
features = [
    'danceability',
    'energy',
    'valence',
    'tempo',
    'loudness',
    'acousticness',
    'speechiness',
    'instrumentalness'
]

df = df[features].dropna()
df.head()
```

```python
def label_mood(row):
    if row['valence'] >= 0.7 and row['energy'] >= 0.6:
        return 'Happy'
    elif row['energy'] >= 0.75:
        return 'Energetic'
    elif row['valence'] < 0.4 and row['energy'] < 0.5:
        return 'Sad'
    else:
        return 'Calm'
```

```python
df['mood'] = df.apply(label_mood, axis=1)
df['mood'].value_counts()
```

```python
import os

os.makedirs("../data/processed", exist_ok=True)
```

```python
df.to_csv("../data/processed/spotify_labeled.csv", index=False)
print("Dataset saved to data/processed/spotify_labeled.csv")
```
